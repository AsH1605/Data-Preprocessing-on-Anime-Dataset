```python
# Equal Width Discretization
num_bins = 5
min_value = df['Episodes'].min()
max_value = df['Episodes'].max()
bin_width = (max_value - min_value) / num_bins
bin_edges = [min_value + i * bin_width for i in range(num_bins + 1)]
bin_labels = [f'Bin {i}' for i in range(1, num_bins + 1)]
df['discretized_column'] = pd.cut(df['Episodes'], bins=bin_edges, labels=bin_labels, include_lowest=True)
df.to_csv('discretized_data.csv', index=False)
plt.hist(df['discretized_column'], bins=num_bins, edgecolor='k', alpha=0.7)
plt.xlabel('Discretized Values')
plt.ylabel('Frequency')
plt.title('Histogram of Discretized Episodes Column')
plt.show()
```

```python
# Outlier Trimming
Q1 = df['Rating'].quantile(0.25)
Q3 = df['Rating'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Rating'] < lower_bound) | (df['Rating'] > upper_bound)]
data = df[(df['Rating'] >= lower_bound) & (df['Rating'] <= upper_bound)]
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df['Rating'], vert=False)
plt.title('Before Outlier Trimming')
plt.xlabel('Rating')

plt.subplot(1, 2, 2)
plt.boxplot(data['Rating'], vert=False)
plt.title('After Outlier Trimming')
plt.xlabel('Rating')

plt.tight_layout()
plt.show()
```

```python
# Outlier Capping
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.boxplot(df['Rating'], vert=False)
plt.title('Before Outlier Capping')

df.loc[df['Rating']  < lower_bound, 'Rating'] = lower_bound
df.loc[df['Rating']  > upper_bound, 'Rating'] = upper_bound

plt.subplot(1, 2, 2)
plt.boxplot(df['Rating'], vert=False)
plt.title('After Outlier Capping')

plt.tight_layout()
plt.show()
```
